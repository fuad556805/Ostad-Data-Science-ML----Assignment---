Problem: 01
write  a python program that takes a number as input and checks whether it is even or odd.

```python
num = int(input("Enter a number: "))
if num % 2 == 0:
    print(f"{num} is even.")
else:
    print(f"{num} is odd.")
```

Problem: 02
write a program that takes two numbers and an operator(+,-,*,/) and performs thae calculation.

```python
num1 = int(input("Enter the first number: "))
operator = input("Enter the operator: ")
num2 = int(input("Enter the second number: "))
result = 0

if operator == '+':
    result = num1+num2
elif operator == '-':
    result = num1-num2
elif operator == '*':
    result = num1*num2
elif operator == '/':
    result = num1/num2
else:
    result = 'Please enter valid (+,-,*,/) operator.'

print(f"{num1}{operator}{num2} = {result}")
```

Problem: 03
write a program using a for loop that calculates the sum of even numbers between 1 and 100.

```python
sum = 0

for i in range(2,101,2):
    sum = sum + i

print(f"Sum of even numbers(1 to 100): {sum}")
```
