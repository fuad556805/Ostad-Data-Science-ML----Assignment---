**14. What is the main difference between `linspace()` and `arange()`?**

```python
# Please consult the lectures. 
# arange() takes 3rd argument as step size.
# linspace() take 3rd argument as no of point we want.


# linspace() creates a fixed number of equally spaced values between a start and end point (including the end).
# arange() creates values using a fixed step size and usually excludes the stop value.

```
